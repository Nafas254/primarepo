```python
import pandas as pd

df=pd.read_csv("energia.csv")
df
```

```python
Territori=df['Territorio'].unique().tolist()
Territori
```

```python
dict_territori={
 "Valle d'Aosta / VallÃ©e d'Aoste":"Valle d Aosta",
 'Trentino Alto Adige / SÃ¼dtirol':'Trentino Alto Adige',
 'Provincia Autonoma Bolzano / Bozen':'Provincia Autonoma Bolzano',
 'Provincia Autonoma Trento':'Provincia Autonoma Trento',
 'Friuli-Venezia Giulia':'Friuli Venezia Giulia','Nord-ovest':'Nord Ovest','Emilia-Romagna':'Emilia Romagna','Nord-est':'Nord est'}
df.Territorio
```

```python
df=df.replace({'Territorio':dict_territori})
Territori=df['Territorio'].unique().tolist()
Territori
```

```python
for i in Territori:
    df.loc[df.Territorio==i,:].to_csv(f'CSV_files/{i}.csv', index=False)
    print(df.loc[df.Territorio==i,:])
```

```python
for i in Territori:
    df.loc[df.Territorio==i,:].to_excel(f'xlsx_files/{i}.xlsx', index=False)
    print(df.loc[df.Territorio==i,:])
```
